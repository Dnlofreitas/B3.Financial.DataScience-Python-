```python
#Biblioteca Data Science
import fundamentus
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#DataFrame B3 Bolsa Brasileira de SP
df = fundamentus.get_resultado()
```

```python
df
```

```python
#Screenning
#Empresas B3 com PL maior que 0 e Margem Ebitda maior 40%
filtro1 = df[ (df.pl > 0) & (df.mrgebit > 0.4) ]
```

```python
#Ordenar os Papeis f1
filtro1 = filtro1.sort_values('dy', ascending = False)
```

```python
#Plotar Grafico
plt.figure (figsize = (20,6));
plt.bar(filtro1.index, filtro1.dy);
plt.xticks(rotation = 90);
plt.title("Empresas B3 com PL maior que 0 e Margem Ebitda maior 40%")
```

```python
#Screenning2
#Empresas B3 com PL maior que 0 e Margem Ebitda maior 40%
filtro2 = df[ (df.pl > 0) & (df.mrgebit > 0.4) & (df.dy > 0.1)]
```

```python
#Ordenar os Papeis f2
filtro2 = filtro2.sort_values('dy', ascending = False)
```

```python
#Plotar Grafico2
plt.figure (figsize = (20,6));
plt.bar(filtro2.index, filtro2.dy);
plt.xticks(rotation = 90);
plt.title("Empresas B3 com PL maior que 0 e Margem Ebitda maior 40% e Divident Yeld")
```
